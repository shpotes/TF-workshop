```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import preprocessing
```

```python
import pathlib
import json
import pickle
import pandas as pd
from tqdm import tqdm
from datetime import datetime
```

```python
from datagen import *
```

```python
metadata = pd.read_csv('data/metadata.csv')
```

```python
train_source = build_source_from_metadata(metadata,
                                          data_dir='data',
                                          mode='train')
val_source = build_source_from_metadata(metadata,
                                        data_dir='data',
                                        mode='val')
```

```python
with open('data/tokenizer.pkl', 'rb') as handle:
    tokenizer = pickle.load(handle)
    
tokenizer
```

```python
NUM_EPOCHS = 1000
BATCH_SIZE = 64
```

```python
train_ds = make_dataset(train_source, 
                        tokenizer,
                        training=True,
                        batch_size=BATCH_SIZE,
                        num_epochs=NUM_EPOCHS)

val_ds = make_dataset(val_source, 
                      tokenizer,
                      training=False,
                      batch_size=BATCH_SIZE)
```

```python
img_input = tf.keras.Input((224, 224, 3), name='img_input')
img_backbone = tf.keras.applications.VGG16(input_tensor=img_input, weights='imagenet', include_top=False)
img_feat = layers.Flatten()(img_backbone.output)
```

```python
question_input = tf.keras.Input(24, name='question_input')
q_emb = layers.Embedding(input_dim=5000, output_dim=64)(question_input)
q_av = layers.GlobalAveragePooling1D(name='question_feat')(q_emb)
# qAv
q_av
```

```python
ans_input = tf.keras.Input((20, 4), name='ans1_input')

ans_emb = layers.Embedding(input_dim=5000, output_dim=32, name='answer_embedding')(ans_input)
ans_av = layers.Lambda(lambda inp: tf.reduce_mean(inp, axis=1), name='word_average')(ans_emb)
ans_av = layers.Flatten()(ans_av)
ans_av
```

```python
act = layers.concatenate([img_feat, q_av, ans_av], name='VQA_feat')
h1 = layers.Dense(512, activation='relu')(act)
answer = layers.Dense(4, activation='softmax', name='answer')(h1)
q_type = layers.Dense(6, activation='softmax', name='type')(h1)
```

```python
model = tf.keras.Model(inputs=[img_input, question_input, ans_input],
                      outputs=[answer, q_type])
```

```python
tf.keras.utils.plot_model(model)
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer=tf.keras.optimizers.Adam(3e-4),
              metrics=['accuracy'])
```

```python
now = datetime.now().strftime("%m/%d/%Y, %H:%M:%S")

tb = tf.keras.callbacks.TensorBoard(log_dir='logs/test/' + now)

steps_per_train = len(train_source) // BATCH_SIZE
steps_per_val = len(val_source) // BATCH_SIZE

model.fit_generator(train_ds, 
                    steps_per_epoch=steps_per_train,
                    validation_data=val_ds,
                    validation_steps=steps_per_val,
                    callbacks=[tb])
```
